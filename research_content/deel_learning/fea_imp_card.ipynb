```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 设置绘图区域背景为白色
plt.rcParams['figure.facecolor'] = 'white'
import seaborn as sns

# 定义一个函数来绘制特征重要性的条形图
def plot_feature_importances(fea_imp, fn, ax, title):
    mean_abs_imp = np.abs(fea_imp).mean(axis=0)
    n = 30  # 选择前30个最重要的特征
    top_indices = np.argsort(-mean_abs_imp)[:n]
    feature_names = [fn[i] for i in top_indices]
    sns.barplot(x=mean_abs_imp[top_indices], y=feature_names, ax=ax)
    ax.set_title(title, fontsize=16)  # 放大标题字体
    ax.set_xlabel('Feature Importance (x1e-3)', fontsize=14)  # 放大x轴标签字体
    ax.set_ylabel('Features', fontsize=14)  # 放大y轴标签字体
    ax.grid(True, which='both', linestyle='--', linewidth=0.5, color='lightgrey')
    # 将横坐标的值乘以1000，保留两位小数
    ax.set_xticklabels([f"{x*1000:.2f}" for x in ax.get_xticks()], fontsize=12, rotation=45)  # 放大刻度标签字体
```

```python
fn = pd.read_csv("/home/luojiawei/EMR_LIP/变量名.csv",header=0).iloc[:,0]
```

```python
fea_imp_pre = np.load("/home/luojiawei/EMR_LIP/结果文件夹/fea_imp_folder/fea_imp_lstm_eicu_death_24h_dym.npy")

# 设置图形大小和布局
fig, axs = plt.subplots(1, 1, figsize=(6, 10))
# 绘制每个模型的特征重要性
plot_feature_importances(fea_imp_pre, fn, axs, 'eICU-CRD')

plt.tight_layout()
plt.show()
```

```python
fea_imp_pre = np.load("/home/luojiawei/EMR_LIP/结果文件夹/fea_imp_folder/fea_imp_lstm_mimic_death_24h_dym.npy")

# 设置图形大小和布局
fig, axs = plt.subplots(1, 1, figsize=(6, 10))
# 绘制每个模型的特征重要性
plot_feature_importances(fea_imp_pre, fn, axs, 'MIMIC-IV')

plt.tight_layout()
plt.show()
```
